```python
def row_correlation(df, top_n=10):
    target_row=df.loc[int(input())]
    correlations=df.apply(lambda row: row.corr(target_row), axis=1)
    top10=correlations.abs().nlargest(top_n)
    return correlations.loc[top10.index]
```
