```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 


pre=pd.read_csv('pretourney.csv')
pre=pre.drop(columns='Unnamed: 0', axis=1)
```

```python
pre.head()
```

The 2001 dataset has the 'Seed' column capitalized. I am going to re-insert the row, concatenating on the common column.

```python
pre=pre[pre['Season'] != 2024]
```

```python
pre.drop(columns='Seed',axis=1, inplace=True)
pre.head()
```

```python
pre1=pd.read_csv('s2.csv')
pre1=pre1.rename(columns={'Seed':'seed'})
pre1.head()
```

```python
total=pd.merge(pre,pre1,how='outer')
```

```python
total.head()
```

```python
total.info()
```

```python
total['Team/Year']=total['Season'].astype(str) + ' ' +total['TeamName']
total.drop(columns=['Season', 'TeamName'], axis=1, inplace=True)
```

```python
total=total[total['Tempo'] !=0]
```

```python
popper=total.pop('Team/Year')
total.insert(0,'Team/Year',popper)
```

```python
total.to_csv('pretourney_clean.csv')
```

```python
tourneyteams=total.dropna()
```
